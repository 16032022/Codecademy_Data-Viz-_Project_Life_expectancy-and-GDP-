```python
# Load the required library from Python to explore the variables that are enclosed in dataset 
import pandas as pd
dataset = pd.read_csv('all_data.csv')
gdp_frame_or = pd.DataFrame(dataset)
gdp_frame_or.columns.values  
```

```python
# PREREQUISITE_1 > Rename variable 'Life expectancy at birth variable' as LE 
gdp_frame=gdp_frame_or.copy()  # Create copy of DataFrame
gdp_frame = gdp_frame.rename(columns = {"Life expectancy at birth (years)": "LE"})  
print(gdp_frame.head())    #Get view of the first items of the dataset 
```

```python
# PREREQUISITE_2 > Load the required libraries from Python to perform summary statistics and visualisation plots 
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
# SHOW SUMMARY STATISTICS AND VISUALLY EXAMINE THE DATA 

#Summary statistics: Year, Life_expectancy and GDP _ quantitative variables
print(gdp_frame.describe()) #n obs = 96  from 2000 till 2015 
#Get only the vaules of interest (mean, median, max and min of LE and GDP)
print(np.mean(gdp_frame.GDP))   #meanGDP = 3880498570768.396
print(np.median(gdp_frame.GDP)) #medianGDP= 1280220000000.0
print(np.min(gdp_frame.GDP))    #minGDP=4415702800.0
print(np.max(gdp_frame.GDP))    #maxGDP = 18100000000000.0
print(np.mean(gdp_frame.LE))    #meanLE=72.78958333333335
print(np.median(gdp_frame.LE))  #medianLE = 76.75
print(np.min(gdp_frame.LE))     #maxLE =44.3
print(np.max(gdp_frame.LE))     #min = 81.0
    
print(gdp_frame['Year'].value_counts()) #For each year the same number of obs (6) are available

##Summary statistics:COUNTRY _ categorical variable 
print(gdp_frame['Country'].value_counts()) #The number of obs for each country is the same and it is equal to 16. 
print(gdp_frame['Country'].value_counts(normalize=True)) #The 6 countries are Chile, China, Germany, Mexico, USA, Zimbabwe. 
 
```

```python
#Is there a correlation between GDP and LE by country?
import scipy.stats
#Chile
data_chile = gdp_frame.loc[['Chile']]
print(data_chile['GDP'].corr(data_chile['LE'])) #Pearson 
print(data_chile['LE'].corr(data_chile['GDP']))
result_chile = scipy.stats.spearmanr(data_chile['LE'], data_chile['GDP']) #Spearman
print(result_chile)
#China
data_china = gdp_frame.loc[['China']]
print(data_china['GDP'].corr(data_china['LE'])) #Pearson 
print(data_china['LE'].corr(data_china['GDP'])) 
result_china = scipy.stats.spearmanr(data_china['LE'], data_china['GDP']) #Spearman
print(result_china)
#Germany
data_germany = gdp_frame.loc[['Germany']]
print(data_germany['GDP'].corr(data_germany['LE'])) #Pearson 
print(data_germany['LE'].corr(data_germany['GDP']))
result_germany = scipy.stats.spearmanr(data_germany['LE'], data_germany['GDP']) #Spearman
print(result_germany)
#Mexico
data_mex = gdp_frame.loc[['Mexico']]
print(data_mex['GDP'].corr(data_mex['LE'])) #Pearson 
print(data_mex['LE'].corr(data_mex['GDP']))
result_mex = scipy.stats.spearmanr(data_mex['LE'], data_mex['GDP']) #Spearman
print(result_mex)
#USA
data_usa = gdp_frame.loc[['United States of America']]
print(data_usa['GDP'].corr(data_usa['LE'])) #Pearson 
print(data_usa['LE'].corr(data_usa['GDP']))
result_usa = scipy.stats.spearmanr(data_usa['LE'], data_usa['GDP']) #Spearman
print(result_usa)
#Zimbabwe
data_zim = gdp_frame.loc[['Zimbabwe']]
print(data_zim['GDP'].corr(data_zim['LE'])) #Pearson 
print(data_zim['LE'].corr(data_zim['GDP']))
result_zim = scipy.stats.spearmanr(data_zim['LE'], data_zim['GDP']) #Spearman
print(result_zim)
```
